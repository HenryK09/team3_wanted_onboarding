새로운 컬럼 income_cat 생성

주요 소득 군 설정  
근로:0  
사업:1  
기타:2  
근로==사업:3  
사업==기타:4  
근로==기타:5  
소득 없음:6 

```python
empty = []
count = 0

from tqdm import tqdm
df['income_cat']  = 6
for i in tqdm(df.index, position=0, leave=True):
  if i in emp_list:
    df['income_cat'][i] = 0
    # print(i)
  if i in bus_list:
    df['income_cat'][i] = 1
    # print(i)
  if i in etc_list:
    df['income_cat'][i] = 2
  if i in no_work:
    df['income_cat'][i] = 3
  if i in no_busi:
    df['income_cat'][i] = 4
  if i in no_etc:
    df['income_cat'][i] = 5
  if i in no_money:
    df['income_cat'][i] = 6
```

```python
## 결제 여부 데이터 타입 변경
## 결제 했을 경우 0, 아닐 경우 1

df.query('has_paid == False')
def make_binary(sent):
  if sent is True:
    sent = 0
  else:
    sent = 1
  return sent
```

```python
df['has_paid']= df['has_paid'].apply(lambda x: make_binary(x))
df
```

```python
# gender 컬럼 데이터 값 변환
# 남자는 0, 여자는 1

def make_gen_bin(sent):
  if sent == 'M':
    sent = 0
  if sent == 'F':
    sent = 1
  return sent
df['gender'] = df['gender'].apply(lambda x: make_gen_bin(x))
```

```python
def make_age_group(sent):
  if sent < 20 and sent >=10:
    age_group =   0
  if sent < 30 and sent >=20:
    age_group =   1
  if sent < 40 and sent >=30:
    age_group =   2
  if sent < 50 and sent >=40:
    age_group =   3
  if sent < 60 and sent >=50:
    age_group =   4
  if sent < 70 and sent >=60:
    age_group =   5
  if sent < 80 and sent >=70:
    age_group =   6
  if sent < 90 and sent >=80:
    age_group =   7
  if sent < 100 and sent >=90:
    age_group =   8
  
  return age_group

# 10대 0
# 20대 1
# 30대 2
# 40대 3
# 50대 4
# 60대 5
# 70대 6
# 80대 7
# 90대 8
```

```python
df['age_group'] = df['age'].apply(lambda x : make_age_group(x))
```

## 결제율을 늘릴 수 있는 방법?
1. 환급액이 많을 수록? 결제를 많이 할까?
2. 연령대에 따른 이용율 및 결제율
-  이용 연령대가 주로 20~30대, 실제 결제 인원의 80% 이상.

2. 환급액이 마이너스(-) 인데도 결제 => 왜죠...??
3. 환급 절차가 수수료를 먼저 결제하고 환급을 받기 때문...? => 프로세스를 바꿔야함?

성별데이터 없는 것들 (-) 인 것들 어떻게 할 지 여부


```python
### 새로운 수수료 설정 
## 0이하일 경우 최저 수수료율인 3 %
fee_add = []
for i in tqdm(df['refund']):
  if i >= 2.540000e+05:
    fee = 5
    fee_add.append(fee)
  elif i >= 9.900000e+04:
    fee = 4
    fee_add.append(fee)
  elif i < 0:
    fee = -3
    fee_add.append(fee) 
  else: 
    fee = 3
    fee_add.append(fee)
```

```python
df['fee_rat']  = fee_add
df['fee_fix'] = round(df['fee_rat']/100, 4) * df['refund']
df
```
