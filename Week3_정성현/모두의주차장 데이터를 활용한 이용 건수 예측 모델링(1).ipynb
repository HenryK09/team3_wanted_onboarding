2020.01.01~2020.09.30 : Training Data

2020.10.01~2020.12.31 : Test Data


```python
Train = df.query('DATE <= "2020.09.30"')
Test = df.query('"2020.10.01"<=DATE<="2020.12.31"')
```

```python
# RAM에 과부하가 걸리니 잠시 USER_ID는 제외해두자
Train.drop(['USER_ID'],axis=1, inplace=True)
Test.drop(['USER_ID'],axis=1, inplace=True)
```
